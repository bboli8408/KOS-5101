## 이론적 배경 (Mathematical Background)

데이터 구조 설계 단계. $N$명의 학생이 $K$개의 문항에 응답 행렬 $Y_{N \times K}$.

## 데이터 시뮬레이션 (Data Simulation)
Python을 사용하여 가상의 데이터를 생성합니다.

```python
import numpy as np
import pandas as pd

# 가상의 프로젝트 데이터 생성
n_students = 100
n_items = 20

# True Parameters
true_theta = np.random.normal(0, 1, n_students)
true_beta = np.random.normal(0, 1, n_items)

# Generate Responses (Rasch)
data_matrix = np.zeros((n_students, n_items))
for n in range(n_students):
    for i in range(n_items):
        p = 1 / (1 + np.exp(-(true_theta[n] - true_beta[i])))
        data_matrix[n, i] = np.random.binomial(1, p)

df_project = pd.DataFrame(data_matrix, columns=[f'Q{i+1}' for i in range(n_items)])
```

## 결과 시뮬레이션 및 가시화 (Visualization)
생성된 데이터 또는 분석 결과를 시각화합니다.

```python
import matplotlib.pyplot as plt

# 문항별 정답률 확인
item_means = df_project.mean()
plt.bar(range(1, n_items+1), item_means)
plt.xlabel('Item Number')
plt.ylabel('Proportion Correct')
plt.title('Project Data: Item Difficulty Checks')
plt.show()
```
