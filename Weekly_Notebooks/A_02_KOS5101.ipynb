## Confirmatory Exploratory Factor Analysis

### Confirmatory Factor Analysis (CFA)

**Confirmatory Factor Analysis (CFA)** is a statistical technique used to verify the factor structure of a set of observed variables. Unlike Exploratory Factor Analysis (EFA), where the model structure emerges from the data, CFA requires the researcher to verify a pre-specified theoretical model derived from theory or previous research.

#### The Measurement Model (측정 모형)
In CFA, we define a **Measurement Model** that specifies the relationship between observed variables ($X$) and underlying latent constructs ($\xi$).

The general equation is:
$$ X = \Lambda \xi + \delta $$

Where:
- **$X$** (Observed Variables): The measured indicators (e.g., test items, survey questions).
- **$\xi$** (Latent Factors): The unobserved theoretical constructs (e.g., Intelligence, Anxiety).
- **$\Lambda$** (Factor Loadings): The regression slopes relating $X$ to $\xi$. Represents the strength of the relationship.
- **$\delta$** (Measurement Errors): The unique variance in each indicator not explained by the factor.

#### Key Characteristics
1.  **Hypothesis Driven**: You must specify which items load on which factors before analyzing the data.
2.  **Constraint of Zero Loadings**: Cross-loadings (an item loading on multiple factors) are typically constrained to zero unless specified otherwise.
3.  **Correlated Factors**: Latent factors are usually allowed to covary ($\phi = \text{Cov}(\xi_1, \xi_2)$).

#### Model Fit Indices (모형 적합도)
To evaluate how well the hypothesized model matches the observed data, we check fit indices:

*   **Absolute Fit**:
    *   **$\chi^2$ (Chi-Square)**: Tests exact fit. Use with caution as it is sensitive to large sample sizes (often rejects good models).
    *   **RMSEA (Root Mean Square Error of Approximation)**: Measures badness-of-fit. Values $< 0.06$ are excellent, $< 0.08$ are acceptable.
    *   **SRMR (Standardized Root Mean Square Residual)**: Average standardized residual. Values $< 0.08$ are acceptable.
    
*   **Incremental Fit** (Comparing to a null model):
    *   **CFI (Comparative Fit Index)**: Values $> 0.95$ are excellent, $> 0.90$ are acceptable.
    *   **TLI (Tucker-Lewis Index)**: Similar to CFI, penalizes model complexity. Values $> 0.90$ or $0.95$.
