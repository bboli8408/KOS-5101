# Week 10: IRT 모델 및 시뮬레이션

## 학습 목표
2-Parameter Logistic (2PL) 모델.

## 이론적 배경 (Mathematical Background)

2PL Model:  
$$ P(X_{ni} = 1) = \frac{1}{1 + \exp(-\alpha_i(\theta_n - \beta_i))} $$  
- $\alpha_i$: Item Discrimination (Slope)

## 데이터 시뮬레이션 (Data Simulation)
Python을 사용하여 가상의 데이터를 생성합니다.

```python
import numpy as np

def ir_2pl(theta, alpha, beta):
    return 1 / (1 + np.exp(-alpha * (theta - beta)))

theta = np.linspace(-4, 4, 100)

# 문항 파라미터 [alpha, beta]
items = [[1.0, 0.0], [2.0, 0.0], [0.5, 0.0]]
```

## 결과 시뮬레이션 및 가시화 (Visualization)
생성된 데이터 또는 분석 결과를 시각화합니다.

```python
import matplotlib.pyplot as plt

for i, (a, b) in enumerate(items):
    p = ir_2pl(theta, a, b)
    plt.plot(theta, p, label=f'a={a}, b={b}')

plt.title('ICC for 2PL Model (Varying Discrimination)')
plt.xlabel('Ability')
plt.ylabel('Probability')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
