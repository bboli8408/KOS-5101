## 데이터 시뮬레이션 (Data Simulation)
Python을 사용하여 가상의 데이터를 생성합니다.

```python
import numpy as np
import pandas as pd

# 가상의 성적 데이터 생성 (Simulation)
np.random.seed(42)
students = [f'Student_{i}' for i in range(1, 21)]
scores = np.random.randint(50, 100, 20)
data = pd.DataFrame({'Student': students, 'Score': scores})
```

## 결과 시뮬레이션 및 가시화 (Visualization)
생성된 데이터 또는 분석 결과를 시각화합니다.

```python
import matplotlib.pyplot as plt

# 성적 분포 시각화
plt.figure(figsize=(10, 6))
plt.bar(data['Student'], data['Score'], color='skyblue')
plt.axhline(y=data['Score'].mean(), color='r', linestyle='--', label=f'Mean: {data['Score'].mean()}')
plt.xticks(rotation=45)
plt.title('Simulated Student Scores')
plt.legend()
plt.tight_layout()
plt.show()
```
