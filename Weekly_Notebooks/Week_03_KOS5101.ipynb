## 이론적 배경 (Mathematical Background)

상관관계(Correlation) 분석:  
$$ r = \frac{\sum(X_i - \bar{X})(Y_i - \bar{Y})}{\sqrt{\sum(X_i - \bar{X})^2 \sum(Y_i - \bar{Y})^2}} $$

## 데이터 시뮬레이션 (Data Simulation)
Python을 사용하여 가상의 데이터를 생성합니다.

```python
import numpy as np
import pandas as pd

# 두 변수 간의 상관관계 시뮬레이션
x = np.random.rand(50) * 10
y = 2 * x + 1 + np.random.normal(0, 2, 50) # 선형 관계 + 노이즈
```

## 결과 시뮬레이션 및 가시화 (Visualization)
생성된 데이터 또는 분석 결과를 시각화합니다.

```python
import matplotlib.pyplot as plt

plt.scatter(x, y, alpha=0.7)
plt.title('Correlation Simulation')
plt.xlabel('Variable X')
plt.ylabel('Variable Y')
plt.grid(True)
plt.show()
```
