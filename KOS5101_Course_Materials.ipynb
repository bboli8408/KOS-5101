# Week 5: CTT 및 Guttman 구조 분석

**학습목표 및 주요내용**
고전검사이론(CTT)의 핵심 개념 학습. Guttman 구조를 활용한 응답 패턴의 정성적 시뮬레이션 및 분석 실습.

```python
import numpy as npimport pandas as pd# CTT Basic Statisticsdef calculate_ctt_stats(responses):    # responses: DataFrame (rows=students, cols=items)    item_difficulty = responses.mean()    total_scores = responses.sum(axis=1)    item_discrimination = []    for col in responses.columns:        corr = responses[col].corr(total_scores)        item_discrimination.append(corr)        return item_difficulty, item_discrimination# Guttman Pattern Simulation Placeholderprint("CTT & Guttman Analysis Section")
```
